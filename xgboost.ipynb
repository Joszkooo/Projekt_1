```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('train.csv', error_bad_lines=False)
df.head()
```

```python
df_test = pd.read_csv('test.csv', error_bad_lines=False)
df_test.head()
```

```python
X_train, y_train = df.iloc[:, 1:], df.iloc[:, 0]
```

```python
from xgboost import XGBClassifier
```

```python
xgboost = XGBClassifier()
```

```python
xgboost.fit(X_train, y_train)
```

```python
pred = xgboost.predict(df_test)
```

```python
pred
```

```python
df = pd.DataFrame(pred)
df = df.reset_index()
df.columns = ['ImageId', 'Label']
df.head()
```

```python

df.to_csv('sample_submission.csv', index=False)
```
