```python
X_train, y_train = df.iloc[:, 1:], df.iloc[:, 0]
```

```python
for col in df_test.columns:
    df_test.at[0, col]
```

```python
def get_error(example_row, digit_row):
    errors = []
    for col in df_test.columns:
        error = digit_row[col] - example_row[col]
        errors.append(error)
    return sum(errors)
get_error(df_test.iloc[0], numbers[0])
```

```python
fig, ax = plt.subplots(4, 3)

for i in range(9):
    max_value = {i: get_error(df_test.iloc[i], numbers[i]) for i in range(10)}
    predicted = min(max_value.items(), key=lambda x: x[1])[0]

    arr = np.array(df_test.iloc[i]).reshape(28, 28)
    ax[i % 3, i //3].imshow(arr, cmap='gray')
    ax[i % 3, i //3].set_title(predicted)
    ax[i % 3, i //3].axis('off')
plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
methods = [DecisionTreeClassifier, KNeighborsClassifier, GaussianNB, RandomForestClassifier, GradientBoostingClassifier]
```

```python
pred_output = {}
for method in methods:
    reg = method().fit(X_train, y_train)
    out = reg.predict(df_test)
    pred_output[method.__name__] = out
```

```python
pred_output
```

```python
for key, value in pred_output.items():
    np.savetxt(f'{key}.csv', np.asarray(value, dtype=np.int32))
```

```python
df = pd.DataFrame(pred_output['DecisionTreeClassifier'])
df['ImageId'] = range(1, 28001)
df.columns = ['Label', 'ImageId']
df = df[['ImageId', 'Label']]
df.head()
```
